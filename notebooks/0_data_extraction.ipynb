```python
import re
import json
import uuid
import pdfplumber
```

```python
texts = []

with pdfplumber.open("../data/raw/1.5.2918.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        texts.append(text)
```

```python
law = "".join(texts)
```

```python
parts = re.split("Madde", law)
parts = parts[:139]
```

```python
chunks = []
for i, chunk in enumerate(parts):
    chunk = chunk.strip()
    chunks.append({
        "chunk_id": f"trafik_kanunu_{i+1}",
        "text": chunk,
        "metadata": {
            "madde_no": i + 1,
            "char_count": len(chunk)
        }
    })
```

```python
with open("../data/processed/chunks/trafik_kanunu.jsonl", "w", encoding="utf-8") as f:
    for chunk in chunks:
        f.write(json.dumps(chunk, ensure_ascii=False) + "\n")
```
